```python
import torch
import matplotlib.pyplot as plt

torch.manual_seed(0)
X = torch.randn(100, 2)                           # 100 data points, 2 features each
y = (X[:, 0] * X[:, 1] > 0).long()                # Label = 1 if x1 * x2 > 0

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
plt.title("Toy Binary Classification")
```

```python
import torch.nn as nn
model = nn.Linear(2, 2)  # Input = 2D point, Output = 2 class scores (logits)
```

```python
import torch.nn.functional as F
import torch.optim as optim

optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
for epoch in range(40):
    logits = model(X)                  # forward pass (logits = W x + b)
    loss = F.cross_entropy(logits, y)  # compares logits to true labels

    loss.backward()                    # compute ∂L/∂W and ∂L/∂b
    optimizer.step()                   # update weights and bias
    optimizer.zero_grad()              # clear old gradients

    
    print(f"Epoch {epoch} - Loss: {loss.item():.4f}")
    
```

```python
w = model.weight.data
b = model.bias.data

x_vals = torch.linspace(-3, 3, 100)
y_vals = -(w[0, 0] * x_vals + b[0]) / w[0, 1]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
plt.plot(x_vals, y_vals, 'k--')  # decision boundary
```
